```python
[작성법]
try: 
    예외가 발생할 수 있는 코드 블록
except: 
    예외가 발생했을 때 수행되는 코드
```

```python
try: 
    num1 = int(input("첫번째 정수를 입력: "))
    num2 = int(input("두번째 정수를 입력: "))
        print("num1/num2의 결과: ", num1/num2)
except: 
    print("예외가 발생했습니다.")
```

```python
[작성법]
try: 
    예외가 발생할 수 있는 코드 블록
except [발생예외1]:
    [발생예외1]에 해당되면 수행되는 코드 블록
except [발생예외2]:
    [발생예외2]에 해당되면 수행되는 코드 블록
```

```python
- finally절은 try문 수행 도중 예외 발생 여부에 상관없이 항상 수행
- 보통 finally절은 사용한 리소스를 close해야할 때 많이 사용
```

```python
[작성법]
try:
    예외가 발생할 수 있는 코드블록
except:
    예외가 발생했을 대 수행되는 코드블록
finally: 
    예외 발생과 상관없이 항상 실행되는 코드 블록
```

```python
try:
    f=open('day01.txt', 'w', encoding = "utf-8")
    f.write("day01.txt 입니다")
except EOFError as e:
    print(e)
finally: 
    f.close()
print(f.closed)    
```

```python
try: 
    age= int(input("나이를 입력하세요: "))
    if age<0 or age>150:
        raise Exception("나이는 0이상 150미만으로 입력해주세요")
except Exception as e:
    print(e)
else:
    print("당신은", age, "살이군요!")
finally:
    print("프로그램을 종료합니다")
```

```python
06. 사용자 정의 예외 클래스
```

```python
[작성법]
class 예외클래스명(Exception)
    def __init__(self):
        super().__init__("예외메시지")
```

```python
class AgeError(Exception):
    def __init__(self):
        super().__init__("사람의 나이는 0이상 150미만으로 입력해주세요")
```

```python
try: 
    age= int(input("나이를 입력하세요: "))
    if age<0 or age>150:
        raise Exception("나이는 0이상 150미만으로 입력해주세요")
except AgeError as e:
    print(e)
else:
    print("당신은", age, "살이군요!")
finally:
    print("프로그램을 종료합니다")
```

```python
## 연습문제 02|
- 두 수를 입력받아 나눗셈 수행.
- 0으로 나누면 ZeroDivisionError, 숫자가 아니면 ValueError 처리.
```

```python
try:
    num1=int(input("정수를 입력하세요: "))
    num2=int(input("정수를 입력하세요: "))
    print(num1/num2)
except ZeroDivisionError as e:
    print(e)
except ValueError as e:
    print(e)
```

```python
## 연습문제 03| 
- 사용자가 입력한 숫자에 해당하는 인덱스를 이용해 리스트에서 값을 출력하는 프로그램을 작성
- 사용자가 입력한 숫자가 리스트의 인덱스를 벗어나면 IndexError 예외처리 - "잘못된 인덱스입니다." 출력
- 사용자가 숫자가 아닌 값을 입력하면 ValueError 예외처리 - "숫자만 입력할 수 있습니다." 출력
- my_list = [10,20,30,40,50]
```

```python
try:
    my_list = [10,20,30,40,50]
    print(my_list[7])
except IndexError as e:
    print(e)
except ValueError as e:
    print(e)
```
