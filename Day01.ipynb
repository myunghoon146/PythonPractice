## 03 | 문자열의 슬라이싱(slicing)

슬라이싱: 무언가를 자르다!!

### 01) 문자열 인덱스 
- 문자열은 문자들의 *순서있는 나열*이기 때문에, 각 문자는 번호(인덱스)를 가지고 있음

[문자열 접근 - 인덱스 활용 2가지 방식]
- 양수 인덱스(왼쪽 -> 오른쪽): 0
- 음수 인덱스(오른쪽 -> 왼쪽): -1부터 시작

"파이썬"
0 1 2 
-3 -2 -1 

```python
slice1 = "파이썬 수업 중입니다"
print(slice1)
```

```python
print(slice1[0])
```

```python
#선을 출력하려면?
print(slice1[2])
```

```python

print*(slice1[4])
```

```python
slice2 = "파이썬"
print(slice2[-1])
print(slice2[-2])
print(slice2[-3])
```

### 02) 슬라이싱 연산 

[시작:끝:간격]
[이상:미만]
-시작: 시작인데긋, 생략하면 처음부터 추출
-끝: 종료인덱스, 생략하면 끝까지 추출 (이것 직전까지 슬라이싱해줘)
- 간격: 인덱스의 증감값, 생략하면 1씩 변화

파이썬 
0,1,2

```python
slice2 = "안녕하세요여기는파이썬교실입니다"
print(slice2[0:4])
print(slice2[1:7])
```

```python
print(slice2[1:7:2])
```

```python
slice3 = "파 이 썬 수 업 중 입 니 다"
print(slice3[::2])
```

```python
slice4 = "안녕하세요여기는파이썬교실입니다"
print(slice4[:5])
```

```python
print(slice4[2:])
```

```python
slice5 = "다니입정현이은름이제"
print(slice5[::-1])
```
