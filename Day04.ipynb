```python
1. 인덱싱
- 무언가를 가리킨다.
- 리스트 안의 특정 위치(번호)에 있는 값을 꺼내기
```

```python
a=['가',10, '20']
#인덱싱 0, 1, 2
#                -3,-2,-1
print(a[0])
print(a[1])
print(a[2])
```

```python
2. 슬라이싱 
- 무언가를 자르다
- 리스트의 일부 구간을 잘라내어 새로운 리스트를 만드는 것
```

```python
[시작: 끝: 간격] -> [start: stop: step]
[이상: 미만]
- start: 시작인덱스, 생략하면 처음부터 추출
- stop: 종료인덱스, 생략하면 끝까지 추출
- step: 인덱스의 증감값, 생략하면 1씩 변화
```

```python
#슬라이싱
a = [1,2,3,4,5,6]
print(a[:])
print(a[0:2])
print(a[1:3])
print(a[1:6:2])
```

```python
3. 리스트 누적합
```

```python
arr = [2,4,6,8,10]
sum = 0 
for a in arr:
    sum += a
print(sum)    
```

## 02|리스트의 삽입

```python
- append(): 해당 리스트의 맨 뒤에 괄호 안의 값을 삽입 -> 선택한 리스트 맨뒤\
- insert(): 원하는 위치에 값을 삽입
- extend(): 리스트를 합칩, 확장
```

```python
append() | 리스트 추가 및 확장
[작성법]
변수명.append(값)
```

```python
- 요소 하나 추가하기
```

```python
a=[10, 20, 30]
a.append(40)
print(a)
```

```python
- range함수르 사용하여 리스트 생성하기
```

```python
arr = []
for i in range(2,12, 2):
    arr.append(i)
print(arr)    
```

```python
insert() | 원한느 위치에 값 추가
[작성법]
리스트명.insert(인덱스번호, 값)
```

```python
a = [1,2,3]
a.insert(1,100)
a.insert(3,200)
print(a)
```

```python
extend() | 리스트 확장
[작성법]
리스트명.extend(추가할 리스트의 변수명)
```

```python
a=[10,20,30]
b=[50,60,70]
a.extend(b)
print(a)
```

```python
** append()와 extend()의 차이점
- append: 첨부하다
- extend: 연장하다
```

```python
a=[10,20,30]
b=[50,60,70]
a.append(b)
print(a)
```

```python
- pop() : 리스트 요소 빼버리기
- remove(): 어느 특정값을 지정하고 삭제
- del 리스트명[인덱스]
- clear(): 리스트 요소 전체 삭제
```

```python
pop() | 인덱스를 통해 제거하기
[작성법]
리스트명.pop() -> 가장 마지막에 있는 요소제거
리스트명.pop(인덱스번호)
```

```python
01) 정렬
```

sort(): 리스트에 있는 값을 작은 값부터 차례대로 (오름차순) 정렬
sort(reverse = True): 리스트에 있는 값을 차례대로 (내림차순) 정렬
reverse(): 뒤집기
sorted(): 리스트를 복사해서 정려하여 원본 손상이 되지 않게 새로 정렬 가능!!!

```python
- 문자열 sort 순서: 대문자, 소문자, 한글
```

```python
sort(0) | 오름차순 정렬
[작성법]
리스트명.sort()
```

```python
a = [10,30,25,15,40]
print('a 리스트 출력: ', a)
a.sort()
print(a)
```

```python
## 문자의 경우\
## -대문자, 소문자 한글 순서 정렬
fruits = ['Banana', 'apple','kiwi','포도', 'banana', 'pear']
fruits.sort()
print(fruits)
```

```python
sort(reverse= True) | 내림차순
[작성법]
변수명.sort(reverse=True)
```

```python
a=[10,30,25,15,40]
a.sort(reverse=True)
print(a)
```

```python
reverse() | 반대로 뒤집기
[작성법]
변수명.reverse()
```

```python
a = [10,30,25,16,40]
a.reverse()
print(a)
```

```python
sorted() | 오름차순 정렬, 원본유지
[작성법]
변수명 = sorted(정렬할 리스트명)
```

```python
num = [3.5,2,1,4]
sorted_num = sorted(num)
print('원본: ', num)
print('오름차순정렬: ', sorted_num)
```

```python
### 03) 리스트 복제시 문제점 
- 리스트를 그냥=로 복사하면 새로 만드는게 아니라 같은 리스트를 함께 가리킴
- 그래서 한쪽을 고치면, 다른 쪽도 같이 바뀜 
```

```python
origin = ['A', 'B', 'C']
copied = origin
print(origin)
print(copied)
```

```python
copied[2]='D'
print(copied)
print(origin)
```

```python
## 해결방법 1:  [:]슬라이싱 사용
```

```python
origin = [1,2,3]
copied = origin[:]
copied[0] = 99
print(origin)
print(copied)
```

```python
해결방법2: list()함수 사용
```

```python
origin= [1,2,3]
copied = list(origin)
copied[0] = 99
print(origin)
print(copied)
```

```python
해결방법 3: copy()메서드를 사용해서 리스트 복제
-이렇게 하면 원본과 복사본이 서로 다른 리스트가 되어, 한쪽을 수정해도 다른 쪽은 그대로 유지가능
```

```python
origin = [1,2,3]
copied=origin.copy()
copied[0] = 99
print(origin)
print(copied)
```
