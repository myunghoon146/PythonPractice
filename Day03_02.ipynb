```python
04) BeautifulSoup으로 파싱(분석)
```

```python
- html.parser : 문서를 구조적으로 해석하는 프로그램
- 다른 파서: lxml=> 빠르다, html5lib=>html를 기준으로 정밀하게 분석
```

```python
soup= BeautifulSoup(html, "html.parser")
```

```python
예제1: 제목(h1) 태그 찾기
```

```python
title_tag = soup.find("h1")
print("제목:", title_tag.text)
```

```python
예제2: 목록(li) 전부 출력
```

```python
list_items = soup.find_all("li")
for item in list_items:
    print("목록:", item.text)
```

```python
from bs4 import BeautifulSoup
html = """
<div class = "box">
    <p class = "desc1"> 설명입니다</p>
    <p class = "desc2" 설명입니다2</p>
    <p class = "desc3"> 설명입니다3</p>
    <a href = "https:..example.com" id="link">링크</a>
</div>
"""

soup = BeautifulSoup(html, "html.parser")
```

```python
p_tag = soup.find('p')
print(p_tag.text)
```

```python
print(p_tag.name)
```

```python
print
```

```python
## 모든 p 태그 찾기
p_tag = soup.find_all('p')
for t in p_tag:
    print(t.text)
```

```python
## 모든 attrs태그 찾기
- <p>태그의 모든 속성을 딕셔너리 형태
p_tag = soup.find_all('p')
for tag in p_tag:
    print(tag.attrs)
```

```python
# class 속성의 ㄱ밧을 가져오는 방법
p_tag = soup.find_all('p')
for tag in p_tag:
    print(tag.get('class'))
```
