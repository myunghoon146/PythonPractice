03| 반복문

### 01) while문
- 특정 조건을 만족하는 동안 반복해서 수행
- 반복의 횟수가 사전에 정해지지 않았거나 특정조건이 만족될 때까지 반복해야하는 경우
- 반복문 실행이 종료되면 다시 조건식으로 이동, 조건식의 결곽가 False 임녀 더이상 반복하지 않는다.
**주의사항**
- 무한 루프에 빠지지 않도록 주의!!!
- 루프를 종료할 수 있는 조건을 명확히 하고, break문을 적절히 사용하여 필요할 때 루프를 빠져나올 수 있게 해야함.

### 연습문제03| 입력받은 수까지 합 구하기
- 사용자로부터 자연수 n을 입력받아, 1부터 n까지의 합을 while문으로 구하시오.

```python
n = int(input("자연수를 입력하세요: "))
total = 0
while 0<n:
    total +=n
    n-=1
print(f"1부터 {n}까지의 합: ",total)
```

3. for문과 range()함수

```python
-파이썬에서 숫자의 범위를 만들어주는 함수
```

```python
range(시작, 끝직전, 증감식)
```

-괄호안의 값이 하나인 경우 range(끝직전)값만 인식하게 됨!
- 시작(start): 시작부분, 생략하면 0부터 시작
- 끝(stop): 끝부분, 끝은 포함하지 않고 직전까지만
- 증감심(step): 얼마씩 증가(또는 감소)할지, 생략하면 기본은 +1

[작성법]
1. 끝직전 값만 입력
for 임시변수 in range(끝직전 값):
    반복할 코드
2. 시작, 끝직전 값
for 임시변수 in range(시작값, 끝직전값):
    반복할 코드
3. 증감값 추가
for 임시변수 in range(시작값, 끝직전값, 증감값):
    반복할 코드

```python
for i in range(10):
    print(i)
```

```python
#시작값, 끝직전값
for i in range(3,10):
    print(i)
```

```python
#시작값, 끝직전값, 증감값 입력
for i in range(2, 10, 2):
    print(i)
```

```python
#시작값, 끝직전값, 증감값
#20씩 끊어서 출력 
for k in range(0, 100, 20):
    print(k,end = "  ")
    print("+20 =>")
print("99까지이므로 끝!!!")    
```

```python
for i in range(1,11):
    print(i, end = "   ")
print("반복 끝")    
```
