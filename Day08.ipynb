```python
03) 재귀호출 사용하기
```

```python
def recursive(n):
    if n == 0:
        return 0
    else:
        recursive(n-1)
        print(n)
a = recursive(5)
print("최종반환값:",a)
```

```python
### 팩토리얼이란?
- 자연수 n에 대해 1부터 n까지 모든 자연수를 곱한 값
- n! = n**(n-1)*(n-2)*...*2*1
- 5! = 5*4*3*2*1
- 1! = 1
- 0! = 1(정의에 대해)
```

```python
-재귀 호출을 사용하여 팩토리얼 구현하기
01. 설계
[종료 조건 정의]
- factorial(1) -> 1을 반환
- 재귀호출을 멈추는 역할
[재귀호출 구조]
- n-1을 다음 단계로 전달하여 문제를 점점 줄임 
```

```python
# 함수 설계해보기 
def 함수이름(n):
    #1. 종료 조건
    if 입력된 값이 1인 경우 == 1:
        1을 반환 
    #2. 재귀호출
    else 
    return n * 함수호출(n-1)
# 함수 호출
객체 = 함수이름(n)
```

```python
03| 객체지향(OOP)
- object oriented programming 약자
- 소프트웨어 설계 방식 중 하나로, 프로그램을 객체라는 독립적인 단위로 나누어 작성하는 방법
- 객체 = 데이터(속성) + 기능(메서드)을 하나로 묶은 것
- 코드의 재사용성과 유지보수성을 높이는데 큰 도움
- 핵심 개념은 클래스와 객체
```

```python
01) 핵심 개념
- 클래스(class): 객체를 만들기 위한 설계도 
- 객체(object): 클래스로부터 생성된 실체
- 메서드(Method): 클래스 안에서 정의된 함수
- 상속: 기존 클래스의 속성과 기능을 물려 받아 새로운 클래스 생성
```

```python
02) 객체지향의 사용 이유?
1. 재사용성
- 한 번 만든 클래스를 여러 곳에서 재활용 가능
- 불필요한 코드 중복 감소
2. 유지보수성
- 코드가 모듈화 되어 있어 수정/보완이 쉬움.
- 다른 코드에 영향을 덜 주고 보완 가능
3. 확장성
- 새로운 기능 추가가 용이(-> 상속, 오버라이딩)
- 기존 코드를 수정하지 않고도 프로그램 확장 가능
4. 협업에 유리
- 역할별로 클래스를 나눠서 개발 가능
- 대규모 프로젝트에서 여러 개발자가 동시에 작업하기 편리
```

```python
### 연습문제 \ 
- class명: Person
- 메서드: setdata, print_function, print_job
- setdata메서드: 속성에 직업을 추가하고 "programmer"로 저장
- print_job메서드를 만들고 "직업은 programmer입니다"를 출력하기
```

```python
class Person:
    def setdata(self,name,job):
        self.name=name
        self.job=job
    def print_function(self):
        print(f"이름은 {self.name}이고, 직업은 {self.job}입니다.")
p3 = Person()
p3.setdata("Alice","programmer")
p3.print_function()
```

```python
print(p3.name)
print(p3.job)
```
