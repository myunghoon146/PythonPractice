```python
계산기를 만들고 기능 상속받기
```

```python
class FourCal:   
    def __init__(self, first, second):
        self.first = first
        self.second = second
    #덧셈
    def add(self):
       result=self.first+self.second
       return result        
    #뺄셈
    def minus(self):
        result = self.first-self.second
        return result
    #곱셈
    def multi(self):
        result = self.first*self.second
        return result
    #나눗셈
    def divide(self):
        result = self.first/self.second
        return result
class FourCal2(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
a=FourCal2(3,4)
print(a.add())
print(a.pow())
```
