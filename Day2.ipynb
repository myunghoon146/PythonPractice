03| 관계 연산자

- 두개의 값을 비교하여 그 결과가 참(True)인지 거짓(False)인지 판별하는 연산자

>: 크다
>=: 크거나 같다
<: 작다
<=: 작거나 같다
!=: 같지 않다
==: 같다

```python
num1 = 10
num2 = 20
print(num1> num2)
print(num1 >= num2)
print(num1 < num2)
print(num1 <= num2)
print(num1 != num2)
```

```python
num3=10
print(num3 == 10) 
print(num3 == '10')
```

```python
myAge = int(input("본인의 나이를 입력해주세요! "))
if myAge % 2 == 0:
    print("제 나이는 짝수입니다")
else:
    print("제 나이는 홀수입니다")
```

04| 논리연산자

- 논리적인 연산을 수행하여 참인지 거짓인지를 판단하는 연산자
- 두개 이상의 논리값을 조합하여 새로운 논리 값을 만들 수 있음
- 논리연산자의 종류
    - and: 연산자를 기준으로 왼쪽과 오른쪽 ㄱ밧이 모두 True일때만 True를 반환하고, 나머지는 False를 반환
    - or: 연산자를 기준으로 왼쪽과 오른쪽 값 중 하나라도 Ture이면 True를 반환하고, 둘 다  False일때만 False를 반환
    - not: 뒤에 따라오는 논리값이 True이면 False를 False이면 True를 반환 

### 02) and: a와 b모두  True이면 True, 아니면 False

```python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

03) or: a와 b중 하나라도 True 이면 True, 아니면 False

```python
print(True or Ture)
print(True or False)
print(False or True)
print(False or False)
```

### 04) not: a가 True이면 False를, a가 False이면 True

```python
print(not '')
print(not "hello")
```

```python
print(not [])
print(not [1,2,3])
```

```python
a = [1,2,3]
if not a: # a => False -> True
        print("a는 비어있습니다.")
else:
    print("a에 값이 있습니다.")
```

연산자의 우선순위

-연산자를 두개 이상 함께 사용할 때는 연산의 순서가 정해져 있음
-이 순서를 연산자의 우선순위라고함
-우선순위에 따라 연산이 수행되고 그 결과가 달라질 수 있기 때문에, 개발자에게 연산자의 우선 순위는 매우 중요함
- 연산자 우선순위를 이해하면 코드를 보다 효율적으로 작성할 수 있으며 예상치 못한 결과를 방지할 수 있음
-

- 우선순위
1. 괄호
2. 거듭제곱
3. 곱셈, 나눗셈, 나머지
4. 덧셈, 뺄셈
5. 비교연산자
6. not
7. and
8. or 
9. (복합)대입연산자

- 괄호
- 복잡한 연산을 수행할 때는 우선 순위를 한눈에 어려울 수 있ㅇ므
-> 따라서 먼저 처리해야할 연산을 소괄호로 묶어서 우선순위를 거시적으로 표시하는 것이 좋음
- 괄호 안에 있는 연산은 최우선 순위로, 다른 연산자보다 먼저 처리됨

- 연산자의 결합방향
-> 연산자를 기준으로 결합 방향이 왼쪽에서 오른쪽으로 결합

```python
value = 80-30+10
print(value)
```

```python
value = 80 - 30*10
print(value)
```

-괄호를 사용하여 우선순위와 결합방향을 조절할 수 있음

```python
value = (80-30)*10
print(value)
```
