```python
- format(): 문자열에 값을 삽입할 때 사용
- count(): 특정 문자열이 등장하는 횟수 반환
- find(): 문자열에서 처음 등장하는 위치 반환(없으면 -1)
- index(): find()와 유사하지만, 없을 경우 ValueError 발생
- join(): 문자열 사이에 값을 넣어 iterable(반복가능객체)의 요소들을 연결 
- split(): 특정 구분자를 구준으로 문자열을 나눔
- replace(): 문자열 내 특정 부분을 다른 문자열로 교체
- 불필요한 문자 제거: lstrip(), rstrip(), strip()
- 대소문자 변화 메서드: lower(), upper()
- 기타
```

```python
format(): 문자열에 값을 삽입할 때 사용
- 정렬하는 방법
```

```python
name = "Alice"
age = 25
format_s = "Name: {}, Age: {}".format(name,age)
print(format_s)
```

```python
- format() 정렬
왼쪽 정렬 : <
오른쪽 정렬 : > 
가운데 정렬: ^

-10의 자리의 필드폭(자료를 출력하는데 사용할 전체 칸) 의미 - 
[작성법]
변수명 = {포맷사양}.format("출력할 문자")
[예] {:<10}.format("Left")
[설명]{:포맷사양시작 < 왼쪽정렬 10자리폭}
```

```python
#왼쪽 정렬
left = "{:<10}".format("left")
print(f" '{left}'")
```

```python
#오른쪽 정렬
right = "{:>10}".format("right")
print(f"'{right}'")
```

```python
#가운데 정렬
center = " {:^10}".format("center")
print(f"'{center}'")
```

```python
count(): 특정 요소의 개수를 세는데 사용
[작성법]
결과값을 담을 변수명 = 문자열 변수.count("찾을 문자열 또는 문자")
```

```python
text = "hello world, hello python"
count_hello = text.count('hello')
count_o = text.count('o')
print('hello', count_hello)
print('o',count_o)
```

```python
find(): 특정 문자열을 찾고자 할때 사용
[작성법]
- 변수명 = 문자열변수명.find('찾을글자')
- 변수명 = 문자열변수명.find('찾을글자', 기준인덱스)
- 변수명 = 문자열변수명.find('찾을단어')
*찾는 문자열이 없는 경우 -1반환
```

```python
#문자 찾기
text= "hello world"
index = text.find('l')
print(index)
```

```python
text= "hello world"
index = text.find('l',5)
print(index)
```

```python
text= "hello world"
index = text.find('world')
print(index)
```

```python
불필요한 문자 제거: strip(), lstrip(), rstrip()
```

```python
- strip()
- 앞뒤 공백 삭제, 또는 특정 문자 제거 
```

```python
text =   "     Hello, world!!!       "
result = text.strip()
print(text)
print(result)
```

```python
- lstrip()
- 문자열의 왼쪽 끝에서 공백이나 지정한 문자를 제거
```

```python
# 왼쪽 공백 삭제
text = "      hello world      "
text2 = text.lstrip()
print(f"'{text2}'")
```

```python
text = "xxxhello worldxxx"
ls = text.lstrip('x')
print(f"'{ls}'")
```

```python
- rstrip()
- 문자열의 오른쪽 끝에서 공백이나 지정한 문자를 제거 
```

```python
text = "      Hello World     "
text2 = text.rstrip()
print(f"'{text2}'")
```

```python
#오른쪽 특정 문자 제거
text = "xxxhello worldxxx"
rs = text.rstrip('x')
print(f"'{rs}'")
```
