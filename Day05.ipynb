```python
## 01 | 튜플
- 튜플은 리스트와 거의 유사한 자료형
- 리스트처럼 인덱싱, 슬라이싱 가능
- 튜플은 ()를 사용하여 생성
- 튜플은 리스트보다 생성 속도가 빠르고 메모리를 적게 사용

01) 리스트와 튜플의 차이점
- 튜플은 한 번 만들면 변경할 수 없음(불변, immutable)
- 리스트는 값을 바꾸거나 추가 삭제 할 수 있음(가변, mutable)
- 튜플 -> 아이템 추가, 삭제 수정 불가능!!!
02) 튜플에서 사용할 수 있는 기능:
- 검색: index(), count(), len()
- 포함 여부 확인: in, not in 
- 정렬: sorted()
- 삭제: del
- 사본생성: 리스트와 마찬가지로 튜플도 새로운 변수에 할당하여 사용할 수 있음!!!

03) 튜플에서 사용할 수 없는 기능
삽입: append(), insert(), extend()
수정: 직접수정 불가
삭제: pop(), remove(), clear(), del 변수명[인덱스]
정렬: sort(), reverse()

04) 튜플은 언제 사용?
- 데이터의 개수와 내용이 고정되어 있는 경우
- 값이 변하면 안되는 경우(ex, 좌표, 학번, 요일 등)
```

```python
### 06) 튜플의 슬라이싱
슬라이싱
- [시작: 끝: 간격] -> [start: stop: step]
- [이상: 미만]
- start: 시작인데긋, 생략하면 처음부터 추출
- stop: 종료인덱스, 생략하면 끝까지 추출
- step: 인덱스의 증감값, 생략하면 1씩 변화
```

```python
p = (1,2,3,4,5,6,7)
print(p[1:4])
```

```python
p = (1,2,3,4,5,6,7)
print(p[:2])
```

```python
p = (1,2,3,4,5,6,7)
print(p[1:6:2])
```

```python
**  step이 음수인 경우
- step 방향에 따라 해석이 달라짐
- step양수 -> start 생략 = 0, stop생략 = 끝
- step음수 -> start 생략 = 마지막, stop 생략 = 0번 인덱스
```

```python
p = (1,2,3,4,5,6,7)
print(p[::-1])
```

```python
# start = 4 -> 인덱스 4에서 시작
# stop = 처음 (인덱스 0)까지 간다
# step = -1 -> 뒤로 (거꾸로) 간다.

p=(1,2,3,4,5,6,7)
print(p[4::-1])
```

```python
[작성법]
딕셔너리명.get(key,기본값)
- 딕셔너리명.get(key) => 기본 반환값은 None
- 딕셔너리명.get(key, 기본값)
```

```python
student = {
    'name' : 'Jone',
    'age': 22,
    'major': 'computer science'
    
}
#$ 키가 존재하는 경우
name = student.get('name')
print(name)
# 키가 존재하지 않는 경우 -> None 반환
graduation_year = student.get('id')
print(graduation_year) #
# 키가 존재하지 않는 경우, 기본값 사용
n = student.get('id','Not specified')
print(n)
      
```

```python
- 교집합: A에도 속하고 B에도 속하는 원소들을 골라 넣은 집합
- 합집합: A에 속하거나 B에 속하는 원소들로 이루어진 집합
- 차집합: 한 집합에서 다른 집합 요소를 뺀 새로운 집합을 생성하는 집합 연산\
    A-B
    B-A
```

```python
setA = {10,20,30,40,50,60}
setB = {50,60,70,80,90,100}
#교집합
print(setA & setB) #and
```

```python
#합집합
print(setA | setB)
```

```python
#차집합
print(setA - setB)
print(setB - setA)
```

```python
## intersection() | 교집합
set1 = {1,2,3,4,5}
set2 = {3,4,5,6,7}
set3 = set1.intersection(set2)
print(set3)
```

```python
set1 = {1,2,3,4,5}
set2 = {3,4,5,6,7}
set3 = set1.union(set2)
print(set3)
```

```python
## difference() | 차집합
set1 = {1,2,3,4,5}
set2 = {3,4,5,6,7}
set3 = set1.difference(set2)
print(set3)
```
