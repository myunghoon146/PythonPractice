```python
## 01 | 튜플
- 튜플은 리스트와 거의 유사한 자료형
- 리스트처럼 인덱싱, 슬라이싱 가능
- 튜플은 ()를 사용하여 생성
- 튜플은 리스트보다 생성 속도가 빠르고 메모리를 적게 사용

01) 리스트와 튜플의 차이점
- 튜플은 한 번 만들면 변경할 수 없음(불변, immutable)
- 리스트는 값을 바꾸거나 추가 삭제 할 수 있음(가변, mutable)
- 튜플 -> 아이템 추가, 삭제 수정 불가능!!!
02) 튜플에서 사용할 수 있는 기능:
- 검색: index(), count(), len()
- 포함 여부 확인: in, not in 
- 정렬: sorted()
- 삭제: del
- 사본생성: 리스트와 마찬가지로 튜플도 새로운 변수에 할당하여 사용할 수 있음!!!

03) 튜플에서 사용할 수 없는 기능
삽입: append(), insert(), extend()
수정: 직접수정 불가
삭제: pop(), remove(), clear(), del 변수명[인덱스]
정렬: sort(), reverse()

04) 튜플은 언제 사용?
- 데이터의 개수와 내용이 고정되어 있는 경우
- 값이 변하면 안되는 경우(ex, 좌표, 학번, 요일 등)
```

```python
### 06) 튜플의 슬라이싱
슬라이싱
- [시작: 끝: 간격] -> [start: stop: step]
- [이상: 미만]
- start: 시작인데긋, 생략하면 처음부터 추출
- stop: 종료인덱스, 생략하면 끝까지 추출
- step: 인덱스의 증감값, 생략하면 1씩 변화
```

```python
p = (1,2,3,4,5,6,7)
print(p[1:4])
```

```python
p = (1,2,3,4,5,6,7)
print(p[:2])
```

```python
p = (1,2,3,4,5,6,7)
print(p[1:6:2])
```

```python
**  step이 음수인 경우
- step 방향에 따라 해석이 달라짐
- step양수 -> start 생략 = 0, stop생략 = 끝
- step음수 -> start 생략 = 마지막, stop 생략 = 0번 인덱스
```

```python
p = (1,2,3,4,5,6,7)
print(p[::-1])
```

```python
# start = 4 -> 인덱스 4에서 시작
# stop = 처음 (인덱스 0)까지 간다
# step = -1 -> 뒤로 (거꾸로) 간다.

p=(1,2,3,4,5,6,7)
print(p[4::-1])
```

```python
2. 내림차순
```

```python
number = (5,2,9,1,5,6)
#정렬된 리스트 반환(내림차순)
sorted_number = sorted(number, reverse= True)
print("원본 튜플:", number)
print(" 내림차순 정렬된 리스트: ", sorted_number)
```

```python
number = (5,2,9,1,5,6)
#정렬된 리스트 반환(내림차순)
sorted_number = tuple(sorted(number, reverse= True))
print("원본 튜플:", number)
print(" 내림차순 정렬된 리스트: ", sorted_number)
```

```python
### 10) 튜플 검색하기
-index()
-count()
-len()
```

```python
index() | 지정한 값이 몇 번째 인덱스 속하는지를 알려주는 메서드
```

```python
my_tuple = (10,20,30,20,40)
#값이 처음 나타나는 위치 인덱스 변환
a = my_tuple.index(20)
b = my_tuple.index(20,2)
print("20의 인덱스:", a)
print("20의 인덱스:", b)
```

```python
count() | 개수를 세는 메서드 
```

```python
my_tuple = (10,20,30,40,20,20)
#20의 개수 세기
a = my_tuple.count(20)
print("20의 개수: ", a)
```

```python
print(my_tuple.count(20))
```

```python
len() | 원소가 몇 개 들어있는지 세어줌
```

```python
my_tuple = (10,20,30,40)
new = len(my_tuple)
print("튜플의 길이: ", new)
```

```python
### 11) 튜플의 활용
```

```python
1. for문과 함께 사용하기
```

```python
a = (10,20,30)
for i in a: 
    print(i)
```

```python
2. 언패킹과 함께 사용하기
- 튜플 안에 여러 값이 묶여 있는 경우 for문과 언패킹을 같이 사용 가능
```

```python
arr = [(1,2), (3,4), (5,6)]
for x,y in arr:
    print(x,y)
```

```python
3. enumerate와 함께 사용
- 튜플의 인덱스와 값을 동시에 가져올 수 있도록 해주는 함수
```

```python
t = ('a', 'b', 'c')
for idx, val in enumerate(t, start=1):#index 시작값 입력
    print(idx,val)
```
