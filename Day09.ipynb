```python
## 02 | 상속이란?
- 이미 만들어진 클래스에 다른 기능들을 추가하고 싶을 때 사용가능
- 즉, 부모클래스가 가진 기능들을 모두 물려 받을 수 있음.
```

```python
01) 상속 사용하기
```

```python
[작성법]
class 부모클래스명:
    사용할 메서드1
    사용할 메서드2: 
    ...
class 자식클래스명(부모클래스명):
    사용할 메서드3:
    사용할 메서드4:    
```

```python
class Person:
    def __init__(self, name, age):
    self.name=name
    self.age=age
    def greet(self):
        print(f"Hello. My name is {self.name}")

class Student(Person):
    def__init__(self, name, age, subject):
        super().__init(name,age)
        self.subject = subject
    def study(self):
        print(f"{self.name} 학생은 {self.subject}공부를 합니다.")
s= Student("Alice", 20, "Math")
s.greet()
s.study()
print(s.name)
print(s.age)
print(s.subject)
```

```python
## 03 | 다중상속이란?
- 다중 상속은 말그대로 여러개의 클래스를 상속하는 것을 말함
- 사용방법은 상속할 대상을 ,를 사용하여 나열
```

```python
01) 다중상속 사용하기
```

```python
[작성법]
class 기반클래스이름1:
    ...
class 기반클래스이름2:
    ...
class 파생클래스이름(기반클래스이름1, 기반클래스이름2):    
```

```python
class Animal:
    def bark(self):
        print("bark")
class Mammal:
    def feed(self):
        print("feed")
class Cat(Animal,Mammal):
    def name(self):
        print('Cat')
cat1 = Cat()
cat1.bark()
cat1.feed()
cat1.name()
```

```python
class Person:
    def greeting(self):
        print("안녕하세요")
class Worker:
    def greeting2(self):
        print("저는 일하는 중입니다.")
class Student:
    def study(self):
        print("저는 공부하는 중입니다.")
class Intern(Person,Worker,Student):
    def greeting3(self):
        super().greeting()
        print("저는 인턴입니다")
Jane = Intern()
Jane.greeting3()
Jane.study()
Jane.greeting2()
```

```python
## 04| 오버라이딩
- 무시하다, 우선하다
- 부모클래스의 메서드를 무시하고 새로운 메서드를 만든다!!!
-> 부모클래스의 메서드를 자식클래스에서 재정의 하는 것 
```

```python
class Animal:
    def sound(self):
        print("오버라이딩 전")
        print("멍멍")
class Dog(Animal):
    def sound(self):
        print("오버라이딩 후")
        print("멍멍멍멍멍멍")
class Cat(Animal):
    def sound(self):
        print("야옹")
d = Animal()
d.sound()
d2 = Dog()
d2.sound()
c = Cat()
c.sound()
```

```python
## 연습문제
- 클래스 C와 클래스 Car를 생성
- 클래스 Car는 클래스 C를 상속받음
- 클래스 C에는 생성자를 통해 데이터를 저장하는 기능 생성
- 클래스 Car는 info_print메서드를 생성하고 출력기능을 만든다
- 출력되는 내용 : 바퀴수 4와 가격 1000원
```

```python
[출력내용]
바퀴수 4
가격 1000
# 객체 생성 및 출력
car = Car(4,1000)
car.info_print()
```

```python
class C:
    def __init__(self,tirN, price):
        self.tirN=tirN
        self.price=price
class Car(C):
    def info_print(self):
        print(f"저는 바퀴수가{self.tirN}이고 가격은 {self.price}입니다")
car1 = Car(4,1000)
car1.info_print()
```

```python
### 연습문제 
- 상속의 개념을 활용하여, 기본 인적 사항(이름, 나이)을 가진 Person 클래스를 만들고, 이를 상속받는 Student 클래스를 작성하여 
학생 번호까지 출력하는 기능을 구현하세요. 
```

```python
class Person:
    def __init__(self,name,age,num):
        self.name=name
        self.age=age
        self.num=num
class Student(Person):
    def info_student(self):
        print(f"저는 학생 이름이 {self.name}이며, 나이는 {self.age}이며, 학생번호는{self.num}입니다.")
student1 = Student("이미나", 17, 578)
student1.info_student()
```
